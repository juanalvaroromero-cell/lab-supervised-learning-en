In this lab, we will explore a dataset that describes websites with different features and labels them either benign or malicious . We will use supervised learning algorithms to figure out what feature patterns malicious websites are likely to have and use our model to predict malicious websites.

Your features will be:

+ URL: it is the anonymous identification of the URL analyzed in the study
+ URL_LENGTH: it is the number of characters in the URL
+ NUMBER_SPECIAL_CHARACTERS: it is number of special characters identified in the URL, such as, “/”, “%”, “#”, “&”, “. “, “=”
+ CHARSET: it is a categorical value and its meaning is the character encoding standard (also called character set).
+ SERVER: it is a categorical value and its meaning is the operative system of the server got from the packet response.
+ CONTENT_LENGTH: it represents the content size of the HTTP header.
+ WHOIS_COUNTRY: it is a categorical variable, its values are the countries we got from the server response (specifically, our script used the API of Whois).
+ WHOIS_STATEPRO: it is a categorical variable, its values are the states we got from the server response (specifically, our script used the API of Whois).
+ WHOIS_REGDATE: Whois provides the server registration date, so, this variable has date values with format DD/MM/YYY HH:MM
+ WHOIS_UPDATED_DATE: Through the Whois we got the last update date from the server analyzed
+ TCP_CONVERSATION_EXCHANGE: This variable is the number of TCP packets exchanged between the server and our honeypot client
+ DIST_REMOTE_TCP_PORT: it is the number of the ports detected and different to TCP
+ REMOTE_IPS: this variable has the total number of IPs connected to the honeypot
+ APP_BYTES: this is the number of bytes transfered
+ SOURCE_APP_PACKETS: packets sent from the honeypot to the server
+ REMOTE_APP_PACKETS: packets received from the server
+ APP_PACKETS: this is the total number of IP packets generated during the communication between the honeypot and the server
+ DNS_QUERY_TIMES: this is the number of DNS packets generated during the communication between the honeypot and the server
+ TYPE: this is a categorical variable, its values represent the type of web page analyzed, specifically, 1 is for malicious websites and 0 is for benign websites

```python

# Date Processing
# Converting to datetime objects as requested
features['WHOIS_REGDATE'] = pd.to_datetime(features['WHOIS_REGDATE'], errors='coerce')
features['WHOIS_UPDATED_DATE'] = pd.to_datetime(features['WHOIS_UPDATED_DATE'], errors='coerce')

# Categorical Encoding
# Transforming CHARSET and WHOIS_COUNTRY into numerical dummies
# We do this directly on 'features' to keep all information in one place
features = pd.get_dummies(features, columns=['CHARSET', 'WHOIS_COUNTRY'], drop_first=True)

# Null Value Imputation
# Handling missing values in CONTENT_LENGTH using the median
features['CONTENT_LENGTH'] = features['CONTENT_LENGTH'].fillna(features['CONTENT_LENGTH'].median())

# Create features_clean (Final Numeric Version)
# We drop any non-numeric columns (like datetime) to create the clean training set
features_clean = features.select_dtypes(exclude=['datetime64[ns]', 'object'])

# Check results
print(f"Original features with dates (shape): {features.shape}")
print(f"Clean features for modeling (shape): {features_clean.shape}")
print("\nFirst 5 rows of features_clean:")
features_clean.head()
```

#### Next, evaluate if the columns in this dataset are strongly correlated.

In the Mushroom supervised learning lab we did recently, we mentioned we are concerned if our dataset has strongly correlated columns because if it is the case we need to choose certain ML algorithms instead of others. We need to evaluate this for our dataset now.

Luckily, most of the columns in this dataset are ordinal which makes things a lot easier for us. In the next cells below, evaluate the level of collinearity of the data.

```python
# Your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Filter only existing numerical columns
# This excludes 'URL', 'CHARSET', 'SERVER', and 'WHOIS' columns that are still objects/strings
numeric_features = features.select_dtypes(include=['int64', 'float64'])

# Generate the correlation matrix
# Note: Columns with missing values (like CONTENT_LENGTH) will be handled by .corr() 
# by excluding those specific pairs from the calculation (pairwise deletion)
correlation_matrix = numeric_features.corr()

# Create the heatmap for visualization
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap - Raw Numerical Features")
plt.show()

# Identify high collinearity pairs (> 0.9)
# We flatten the matrix and filter for values above the threshold
high_corr = correlation_matrix.unstack()
high_corr = high_corr[(abs(high_corr) > 0.9) & (high_corr < 1.0)]

print("Highly Correlated Features found in raw data:")
if not high_corr.empty:
    print(high_corr.sort_values(ascending=False).drop_duplicates())
else:
    print("No features with correlation > 0.9 found.")
```

Although there are so many unique values in the `SERVER` column, there are actually only 3 main server types: `Microsoft`, `Apache`, and `nginx`. Just check if each `SERVER` value contains any of those server types and re-label them. For `SERVER` values that don't contain any of those substrings, label with `Other`.

At the end, your `SERVER` column should only contain 4 unique values: `Microsoft`, `Apache`, `nginx`, and `Other`.

```python
# Count `SERVER` value counts here
# Function to group servers into the 4 requested categories
def group_server(server_name):
    # Convert to string and lowercase to easily check for substrings
    server_lower = str(server_name).lower()
    
    # Check for the main server types
    if 'microsoft' in server_lower:
        return 'Microsoft'
    elif 'apache' in server_lower:
        return 'Apache'
    elif 'nginx' in server_lower:
        return 'nginx'
    else:
        # Anything else (including NaNs or weird names) becomes 'Other'
        return 'Other'

# Apply the function to the SERVER column
websites['SERVER'] = websites['SERVER'].apply(group_server)

# Count SERVER value counts here
server_counts_final = websites['SERVER'].value_counts()

print("--- Final SERVER value counts ---")
print(server_counts_final)
```

# Bonus Challenge - Feature Scaling

Problem-solving in machine learning is iterative. You can improve your model prediction with various techniques (there is a sweetspot for the time you spend and the improvement you receive though). Now you've completed only one iteration of ML analysis. There are more iterations you can conduct to make improvements. But you will make constant efforts after the bootcamp to eventually get there.

However, now we do want you to learn one of the advanced techniques which is called *feature scaling*. The idea of feature scaling is to standardize/normalize the range of independent variables or features of the data. This can make the outliers more apparent so that you can remove them. This step needs to happen during Challenge 6 after you split the training and test data because you don't want to split the data again which makes it impossible to compare your results with and without feature scaling. For general concepts about feature scaling, click [here](https://en.wikipedia.org/wiki/Feature_scaling). To read deeper, click [here](https://medium.com/greyatom/why-how-and-when-to-scale-your-features-4b30ab09db5e).

In the next cell, attempt to improve your model prediction accuracy by means of feature scaling. A library you can utilize is `sklearn.preprocessing.RobustScaler` ([documentation](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.RobustScaler.html)). You'll use the `RobustScaler` to fit and transform your `X_train`, then transform `X_test`. You will use logistic regression to fit and predict your transformed data and obtain the accuracy score in the same way. Compare the accuracy score with your normalized data with the previous accuracy data. Is there an improvement?

```python
# Your code here
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# 1. Initialize the RobustScaler
scaler = RobustScaler()

# 2. Fit the scaler on the training data AND transform it
# (We only 'fit' on the training data to prevent data leakage from the test set)
X_train_scaled = scaler.fit_transform(X_train)

# 3. Transform the testing data using the already fitted scaler
X_test_scaled = scaler.transform(X_test)

# 4. Initialize a fresh Logistic Regression model
lr_model_scaled = LogisticRegression(max_iter=1000, random_state=42)

# 5. Fit the model using the SCALED training data
lr_model_scaled.fit(X_train_scaled, y_train)

# 6. Predict the labels for the SCALED testing data
y_pred_scaled = lr_model_scaled.predict(X_test_scaled)

# 7. Calculate the confusion matrix and accuracy score
cm_scaled = confusion_matrix(y_test, y_pred_scaled)
acc_scaled = accuracy_score(y_test, y_pred_scaled)

# 8. Print the results to compare
print("--- Confusion Matrix (Scaled Logistic Regression) ---")
print(cm_scaled)

print("\n--- Accuracy Score ---")
print(f"{acc_scaled * 100:.2f}%")
```

## Comments:

**Did I see an improvement in the accuracy score?**
Counterintuitively, no. The overall accuracy slightly decreased from 92.16% (unscaled Logistic Regression) to 91.60% (scaled with RobustScaler).

**Detailed Analysis of the Confusion Matrix:**
While the accuracy dropped, the distribution of errors shifted:

- False Positives improved: The model became more conservative, only misclassifying 4 safe websites as malicious (down from 7).

- False Negatives worsened: The model missed 26 malicious websites (up from 21).

**Why did this happen?**
In many datasets, outliers are noise. However, in cybersecurity (malicious website detection), extreme values (outliers)—such as an abnormally high CONTENT_LENGTH or unusual packet sizes—are often the exact signals that identify a threat. RobustScaler is specifically designed to minimize the influence of outliers by scaling the data according to the Interquartile Range (IQR). By diminishing the impact of these extreme values, we made the mathematical computation stable (resolving the previous ConvergenceWarning), but we likely blurred the very features that the linear model needed to flag malicious sites effectively.

#### **Conclusion:** Feature scaling is essential for distance-based and linear models to converge properly. However, this iteration proves that for this specific dataset, tree-based models (like our max_depth=5 Decision Tree, which achieved 93.56% accuracy and requires no feature scaling) are natively much better suited to capture the complex, non-linear relationships and valuable outliers inherent in cybersecurity data.
